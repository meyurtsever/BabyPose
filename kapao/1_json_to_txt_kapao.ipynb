```python
import csv
import os
import json
from pprint import pprint
import glob
import re
```

```python
# This notebook combines CSV outputs of KAPAO HPE.
# JSON outputs have excess information for our application.
# We need only detected keypoint coordinates - which are x and y values on the video frames.
# This notebook does exactly this - extracting only x and y values and writing it to another file.
# Later, these seperate txt outputs will be merged for train_test_split to train the model.
```

```python
def write_frames_txt(vid_file_name, kps):
    #os.chdir(root_path)
    print('gelen size: ', len(kps))
    print('gelen size2: ', len(kps[0]))
            
    output_file = str (vid_file_name + ".txt")  
    with open(output_file, "w") as text_file:
        for i in range(len(kps)):
            for j in range(34): 
                text_file.write('{}'.format(kps[i][j]))
                if j < 33: 
                    text_file.write(',')
            text_file.write('\n')
```

```python
frame_kps = []
kps = []

for file in sorted(glob.glob("*.csv"), key = os.path.getmtime):
    
    # read csv file.
    csvreader = csv.reader(open(file))
    rows = []
    for row in csvreader:
        rows.append(row)

    # delete blank rows
    rows = list([x for x in rows if x])
        
    # convert list of lists (string) to list of lists (float)
    frame_list = [list(map(float, sublist)) for sublist in rows]
    
    # each csv contains frame_list of related video clip. we loop over these.
    for frame in frame_list:

        j = 0
        for i in range(34):
            frame_kps.append(frame[j])
            j += 1
            if ((j+1) % 3 == 0):
                j += 1

        kps.append(frame_kps)
        frame_kps = []

    write_frames_txt(file, kps)
    frame_kps = []
    kps = []
```
