### Combine JSON Outputs for AlphaPose

```python
def write_frames_txt(vid_file_name, kps):
    os.chdir(root_path)
    print('gelen size: ', len(kps))
    print('gelen size2: ', len(kps[0]))
            
    output_file = str (vid_file_name + ".txt")  
    with open(output_file, "w") as text_file:
        for i in range(len(kps)):
            for j in range(36): 
                text_file.write('{}'.format(kps[i][j]))
                if j < 35: 
                    text_file.write(',')
            text_file.write('\n')
```

```python
# go to the json directory.
# all of the json outputs are here.
os.chdir(data_path)

kps = []
frame_kps = [] 

# iterate over all json files.
for file in sorted(glob.glob("*.json"), key = os.path.getmtime):
    os.chdir(data_path)   
    
    print ("Reading file: " + str(file))   
    with open(file) as data_file:        
             
        data = json.load(data_file)
        
        # every json file holds keypoint coordinates for all of the frames in video.
        # this 'item' var. will iterate over frames in video.
        for item in data:
            # read alphapose output frame by frame 
            # next, we'll combine them.
            pose_keypoints = data[item]['people'][0]['pose_keypoints_2d']
            
            # delete confidence info. [x, y, c]
            j = 0
            for i in range(36):
                frame_kps.append(pose_keypoints[j])
                j += 1
                if ((j+1) % 3 == 0):
                    j += 1
                    
            # append current frame's output to another list.
            # this list contains all of the outputs of alphapose for its all frames.
            kps.append(frame_kps)
            frame_kps = []
    
        # print('kontrol1: ', len(kps))
        # print('kontrol2: ', len(kps[0]))
    # we got all the keypoints from json.
    # now we will write those combined keypoints to txt file.
    write_frames_txt(file, kps)
    frame_kps = []
    kps = []
```
